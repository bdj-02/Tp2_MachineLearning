## Question 1 : VOLUME

```python
data
```

Lors d’une dépense, combien de produits un client achète-t-il généralement
?

Process : 

Données utilisées 
- Country ==> Pays où l’achat a été réalisé
- Quantity ===> Quantité de produits 
- CustomerID ==> Identifiant de l’acheteur


```python
Volume= data.groupby(['Country', 'CustomerID']).agg({'Quantity':'count'}).reset_index()
```

```python
Volume
```

le nombre moyen
et le nombre médian de produits acheté lors d’une dépense, par pays

```python
Moyenne = data.groupby(['Country']).agg({'Quantity':'mean'}).rename(columns={'Quantity' :'Moyenne'}).reset_index()
```

```python
Moyenne.head(10)
```

```python
Mediane= data.groupby(['Country']).agg({'Quantity':'median'}).rename(columns={'Quantity' :'Médiane'}).reset_index()
```

```python
Mediane.head(10)
```

```python
Moyenne['Médiane'] = Mediane.Médiane
Moyenne
```

```python
sns.displot(Moyenne.Moyenne, kde=False, color = 'c', label = 'Country')
sns.displot(Mediane.Médiane, kde=False, color = 'b', label = 'Country')
```

## Question 2 : MONTANT

Process : 

Données utilisées 
- Quantity : Quantité de produits achetés
- UnitPrice : Prix unitaire du produit
- CustomerID ==> Identifiant de l’acheteur

```python
data.head(5)
```

```python
Montant=data.groupby(['Country']).mean('Quantity').reset_index()
Montant_M=data.groupby(['Country']).median('Quantity').reset_index()
```

```python
Montant.head()
```

```python
Montant['Amount_mean']=(Montant['Quantity'] * Montant['UnitPrice']).round(2)
```

```python
Montant
```

```python
Montant_M.head()
```

```python
Montant_M['Amount_median']=(Montant_M['Quantity'] * Montant_M['UnitPrice']).round(2)
```

```python
Montant_M.head()
```

```python
#Montant['Amount_median'] = Montant_M.Amount_median
#Montant
```

```python
sns.displot(Montant.Amount_mean, kde=False, color = 'c', label = 'Country')
sns.displot(Montant_M.Amount_median, kde=False, color = 'b', label = 'Country')
```
